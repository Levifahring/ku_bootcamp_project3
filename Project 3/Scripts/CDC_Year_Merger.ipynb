```python
import pandas as pd
from pathlib import Path
import datetime
```

```python
raw_df = pd.read_csv(Path("../Resources/Mental_Health_Care_in_the_Last_4_Weeks.csv"))
raw_df.head()
```

```python
del raw_df['Phase']
del raw_df['Time Period']
del raw_df['LowCI']
del raw_df['HighCI']
del raw_df['Confidence Interval']
del raw_df['Quartile Range']
del raw_df['Suppression Flag']
raw_df.head()
```

```python
cdc_national_biweekly = raw_df
cdc_national_biweekly=cdc_national_biweekly[['Indicator', 'Group', 'Subgroup', 'Time Period Label', 'Time Period Start Date', 'Time Period End Date', 'Value']]
cdc_national_biweekly.head()
```

```python
cdc_national_biweekly.to_csv("../Resources/cdc_national_biweekly.csv", index=False, header=True)
```

```python
cdc_statutory_biweekly = raw_df
cdc_statutory_biweekly=cdc_statutory_biweekly[['Indicator', 'State', 'Time Period Label', 'Time Period Start Date', 'Time Period End Date', 'Value']]
cdc_statutory_biweekly.head()
```

```python
cdc_statutory_biweekly.to_csv("../Resources/cdc_statutory_biweekly.csv", index=False, header=True)
```

```python
cdc_annual = raw_df
cdc_annual['Year'] = cdc_annual['Time Period Start Date'].str[-4:].apply(int)
cdc_annual=cdc_annual[['Indicator', 'Group', 'State', 'Subgroup', 'Year', 'Value']]
cdc_annual.head()
```

```python
cdc_statutory_annual = cdc_annual.groupby(["Indicator", "State", "Year"])['Value'].mean().reset_index().sort_values("Year").reset_index()
cdc_statutory_annual = cdc_statutory_annual[["Indicator", "State", "Year","Value"]]
cdc_statutory_annual.head()
```

```python
cdc_statutory_annual.to_csv("../Resources/cdc_statutory_annual.csv", index=False, header=True)
```

```python
cdc_national_annual = cdc_annual.groupby(["Indicator", "Group", "Subgroup", "Year"])['Value'].mean().reset_index().sort_values("Year").reset_index()
cdc_national_annual = cdc_national_annual[["Indicator", "Group", "Subgroup", "Year","Value"]]
cdc_national_annual.head()
```

```python

cdc_national_annual.to_csv("../Resources/cdc_national_annual.csv", index=False, header=True)
```
